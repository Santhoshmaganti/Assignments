```python
friends={"name":("hari","harsha","varun","vamsi","lavanya"),
        "quality":("intelligent","irritating","anxious","happy","topper")}
```

```python
import pandas
import pandas as pd
dff=pd.DataFrame(friends)
dff
```

```python
colleagues={"name":("anusha","jayasree","nasheena","shabana","parvin"),
           "quality":("comedy","intelligent","hardworking","silent","stresshandler")}
```

```python
dfc=pd.DataFrame(colleagues)
dfc
```

```python
df_inner=dfc.merge(dff,how="inner",on="quality")
df_inner
```

```python
df_outer=dfc.merge(dff,how="outer",on="quality")
df_outer
```

```python
df_left=dfc.merge(dff,how="left",on="quality")
df_left
```

```python
df_right=dfc.merge(dff,how="right",on="quality")
df_right
```

```python
dict1={"monthname":("jan","feb","mar","apr"),
      "season":("winter","winter","winter","summer")}
```

```python
df1=pd.DataFrame(dict1)
df1
```

```python
dict2={"monthname":("may","jun","jul","aug"),
      "season":("summer","summer","summer","rainy")}
```

```python
df2=pd.DataFrame(dict2)
df2
```

```python
dict3={"monthname":("sep","oct","nov","dec"),
      "season":("rainy","rainy","rainy","winter")}
```

```python
df3=pd.DataFrame(dict3)
df3
```

```python
df_outer=df1.merge(df2,how="outer",on="season").merge(df3,how="outer",on="season")
df_outer
```

```python
df_inner=df1.merge(df2,how="inner",on="season").merge(df3,how="inner",on="season")
df_inner
```

```python
df_left=df1.merge(df2,how="left",on="season").merge(df3,how="left",on="season")
df_left
```

```python
df_right=df1.merge(df2,how="right",on="season").merge(df3,how="right",on="season")
df_right
```
